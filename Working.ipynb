```python
# This method finds the urls for each of the rosters in the NBA using regexes.
def build_team_urls():
    # Open the espn teams webpage and extract the names of each roster available.
    f = urllib.request.urlopen('https://www.espn.com/nba/teams')
    teams_source = f.read().decode('utf-8')
    teams = dict(re.findall("www\.espn\.com/nba/team/_/name/(\w+)/(.+?)\",", teams_source))
    # Using the names of the rosters, create the urls of each roster
    roster_urls = []
    for key in teams.keys():
        # each roster webpage follows this general pattern.
        roster_urls.append('https://www.espn.com/nba/team/roster/_/name/' + key + '/' + teams[key])
        teams[key] = str(teams[key])
    return dict(zip(teams.values(), roster_urls))
```

```python
def calculate_average(individual_info,individual_info2):
    #skip gs as we drop, 1

    #min played
    individual_info[2] = ((individual_info[2]*individual_info[0]) + (individual_info2[2]*individual_info2[0]))/(individual_info[0] + individual_info2[0])

    #makes the total fgm
    individual_info[3] = (individual_info[3]*individual_info[0]) + (individual_info2[3]*individual_info2[0])
    #makes total fga
    individual_info[4] = (individual_info[4]*individual_info[0]) + (individual_info2[4]*individual_info2[0])

    #fg%
    individual_info[5] = individual_info[3]/individual_info[4]

    #updates fgm to a per game basis
    individual_info[3] = individual_info[3]/(individual_info[0] + individual_info2[0])
    #updates fga to a per game basis
    individual_info[4] = individual_info[4]/(individual_info[0] + individual_info2[0])

    #3pm
    individual_info[6] = (individual_info[6]*individual_info[0]) + (individual_info2[6]*individual_info2[0])
    #3pa
    individual_info[7] = (individual_info[7]*individual_info[0]) + (individual_info2[7]*individual_info2[0])
    #3p%
    if (individual_info[6] > 0.1 and individual_info[7] > 0.1):
        individual_info[8] = individual_info[6]/individual_info[7]

    #get percentage of 3pm
    individual_info[6] = individual_info[6]/(individual_info[0] + individual_info2[0])
    #get percentage of 3pa
    individual_info[7] = individual_info[7]/(individual_info[0] + individual_info2[0])

    #ftm
    individual_info[9] = (individual_info[9]*individual_info[0]) + (individual_info2[9]*individual_info2[0])
    #fta
    individual_info[10] = (individual_info[10]*individual_info[0]) + (individual_info2[10]*individual_info2[0])
    #ft%
    individual_info[11] = individual_info[9]/individual_info[10]

    #get percentage of ftm
    individual_info[9] = individual_info[9]/(individual_info[0] + individual_info2[0])
    #get percentage of fta
    individual_info[10] = individual_info[10]/(individual_info[0] + individual_info2[0])



    #offensive rebounds
    individual_info[12] = ( (individual_info[12]*individual_info[0]) + (individual_info2[12]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #defensive rebounds
    individual_info[13] = ((individual_info[13]*individual_info[0]) + (individual_info2[13]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #total rebounds
    individual_info[14] = individual_info[12] + individual_info[13]

    #assists
    individual_info[15] = ((individual_info[15]*individual_info[0]) + (individual_info2[15]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #blocks
    individual_info[16] = ((individual_info[16]*individual_info[0]) + (individual_info2[16]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #steal
    individual_info[17] = ((individual_info[17]*individual_info[0]) + (individual_info2[17]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #points
    individual_info[20] = ((individual_info[20]*individual_info[0]) + (individual_info2[20]*individual_info2[0])) /(individual_info[0] + individual_info2[0])


    #total games updated correctly
    individual_info[0] += individual_info2[0]
    return individual_info
```
